```python
#X_df contains all the X features
Y_df=df[['SalePrice']]
#Y_df contains all the Y features
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df,Y_df)
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
e=np.asanyarray(X_train)
v=np.asanyarray(y_train)
regr.fit(e,v)
```

```python
w=np.asanyarray(X_test)
l=np.asanyarray(y_test)
l_hat=regr.predict(w)
```

```python
print("Residual sum of squares: %.2f"% np.mean((l_hat - l) ** 2))
from sklearn.metrics import r2_score
r2=r2_score(l,l_hat)
print('R-square Score:', r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

XXX_DATA=PolynomialFeatures(degree = 2).fit_transform(e)
model = LinearRegression()
model.fit(XXX_DATA,v)

XXX1_DATA=PolynomialFeatures(degree = 2).fit_transform(w)
YYY1_DATA=model.predict(XXX1_DATA)
```

```python
r2 = r2_score(YYY1_DATA,l)
print('R-square Score:', r2)

#an average r2 score for polynomial regression for training data as divided before.But it is quite better than the linear
#regression.But we have a pretty bad linear algo working. S the data is a polynomial regression and hence it is done
```
